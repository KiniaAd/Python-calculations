```python
# Import the modules
import matplotlib
from matplotlib import pyplot as plt
import numpy as np
```

```python
# General features
N_sample_points = 25
x_min = -(2*np.pi)
x_max = 3*np.pi
x_label = 'x / rad'
y_min = 1.5
y_max = -1.5
y_label = 'f(x)'
sin_line_style = '-'
cos_line_style = '--'
sin_line_color = 'red'
cos_line_color = 'blue'
linewidth = 2
legend_position = 'upper left'
```

```python
# Calculate the sinus and cosinus functions for x_values
x_values = np.linspace(start = x_min,
                       stop = x_max,
                       num = N_sample_points)
y_sin = np.sin(x_values)
y_cos = np.cos(x_values)
```

```python
# Setup the plot
```

```python
# Add a grid 
plt.grid(axis='both', 
         color="grey",
         linestyle="--",
         alpha=0.5, 
         linewidth=0.75)

# Plot the sinus
plt.plot(x_values,
        y_sin,
        label="sin(x)",
        marker='o',
        markerfacecolor='green',
        markeredgecolor='green',
        ms=5,
        color = sin_line_color,
        linestyle = sin_line_style,
        linewidth = linewidth)

# Plot the cosinus
plt.plot(x_values,
        y_cos,
        label="cos(x)",
        markerfacecolor='purple',
        markeredgecolor='purple',
        marker='^',
        ms=7,
        color = cos_line_color,
        linestyle = cos_line_style,
        linewidth = linewidth)

# Add legend
plt.legend(loc=legend_position)

# Set up the limits of the plot
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)

# Add the labels of the axes
plt.xlabel(x_label)
plt.ylabel(y_label)

# Use custom labels for the x axis
plt.xticks(np.linspace(-2*np.pi, 3*np.pi, 9),[r'$-2\pi$', r'$-2\pi/3$',r'$-\pi$', r'$-\pi/2$','0',r'$\pi/2$',r'$\pi$',r'$2\pi/3$',r'$2\pi$'])

# Plot the black frame 
plt.gca().spines['top'].set_linewidth(1)  # Top border
plt.gca().spines['right'].set_linewidth(1)  # Right border
plt.gca().spines['bottom'].set_linewidth(1)  # Bottom border
plt.gca().spines['left'].set_linewidth(1)  # Left border
plt.gca().spines['top'].set_color('black')  # Top border color
plt.gca().spines['right'].set_color('black')  # Right border color
plt.gca().spines['bottom'].set_color('black')  # Bottom border color
plt.gca().spines['left'].set_color('black')  # Left border color

# Show the plot
plt.show()
```

```python
# Plot
plt.figure(figsize=(8, 6))
plt.imshow(z_2darray, extent=(x_min, x_max, x_min, x_max), origin='lower', cmap='viridis', vmin=-2, vmax=2)
plt.colorbar(label='Z')
plt.xlabel('x / rad')
plt.ylabel('y / rad')
plt.contourf(x_2darray, y_2darray, z_2darray, cmap='viridis')
plt.grid(False)
plt.show()
```
